```python
import yfinance as yf
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period = 'max')
tesla_data.reset_index(inplace=True)

tesla_data.head()
```

```python
# Question 2: Use Webscraping to Extract Tesla Revenue Data
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html5lib")
tesla_revenue=pd.read_html(url, match="Tesla Quarterly Revenue", flavor='bs4')[0]
tesla_revenue.head()
```

```python
# Question 3: Use yfinance to Extract Stock Data
gme = yf.Ticker('GME')
gme_data = gme.history(period = 'max')
gme_data.head()
```

```python
# Question 4: Use Webscraping to Extract GME Revenue Data


url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
tablas = pd.read_html(response.text)
gme_revenue = tablas[0]
gme_revenue.tail()
```

```python
# Question 5: Plot Tesla Stock Graph
import pandas as pd
import matplotlib.pyplot as plt
def make_graph(data, title='Tesla Stock Price'):
    plt.figure(figsize=(10,6))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
make_graph(tesla_data, title='Tesla Stock Price Over Time')
```

```python
# Question 6: Plot GameStop Stock Graph
def make_graph(data, title='GameStop Stock Price'):
    plt.figure(figsize=(10,6))
    plt.plot(data.index, data['Close'], label='GameStop Stock')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
make_graph(gme_data, title='GameStop Stock Price Over Time')

```
